```python
from skimage.feature import hog
from skimage.transform import resize
```

```python
folder_dogs='Train/DOG'
```

```python
import cv2
```

```python
import os
image_paths = [os.path.join(folder_dogs, fname)
               for fname in os.listdir(folder_dogs)
               if fname.lower().endswith(('.jpg', '.jpeg', '.png'))

               ]
#gettting path name from each image so we can use cv2.imread to get the image and use opencv to transform
```

```python
x=[]
y=[]
for image_path in image_paths:
    image=cv2.imread(image_path)
    image_gs=cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    resized_image=resize(image_gs, (64, 128))
    features=hog(resized_image, pixels_per_cell=(8, 8), cells_per_block=(2, 2), visualize=False)
    x.append(features) #creating a list x which is a 2nd order matrix where we keep appending a list of numbers from hog so each line represents one image
    y.append(1)#this list denotes the label  so 1 represents dog
    
#using sklearns hog feature to convert image to numeric data
 
```

```python
folder_cat='Train/CAT'
```

```python
image_cat_path=[os.path.join(folder_cat, fname)
                for fname in os.listdir(folder_cat)
                if fname.lower().endswith(('.jpg'))
               ]
```

```python

for path in image_cat_path:
    image=cv2.imread(path)
    image_gs=cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    resized=resize(image_gs, (64, 128))
    features=hog(resized, pixels_per_cell=(8, 8), cells_per_block=(2, 2), visualize=False)
    x.append(features)
    y.append(0)
    #same as before but 0 is label for cat
```

```python
x=np.array(x)
y=np.array(y)
#converting the lists to numpy arrays
```

```python
from sklearn.utils import shuffle
x, y= shuffle(x, y, random_state=32)
#shuffling the dataset with corresponding x and y together so model learns better
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test=train_test_split(x, y, test_size=0.2)
#splitting 20% of data for testing 
```

```python
class SVM:

    def __init__(self, learning_rate=0.001, lambda_param=0.01, iters=100):
        self.lr = learning_rate
        self.lambda_param = lambda_param
        self.iters = iters
        self.w = None
        self.b = None

    def fit(self, x, y):
        n_samples, n_features = x.shape

        y_ = np.where(y <= 0, -1, 1)

        self.w = np.zeros(n_features)
        self.b = 0

        for _ in range(self.iters):
            for idx, x_i in enumerate(x):
                condition = y_[idx] * (np.dot(x_i, self.w) - self.b) >= 1
                if condition:
                    self.w -= self.lr * (2 * self.lambda_param * self.w)
                else:
                    self.w -= self.lr * (2 * self.lambda_param * self.w - np.dot(x_i, y_[idx]))
                    self.b -= self.lr * y_[idx]


    def predict(self, x):
        approx = np.dot(x, self.w) - self.b
        if approx<0:
            return 0;
        else:
            return 1;
        #creation of own SVM model using numpy
```

```python
model=SVM()
```

```python
model.fit(x_train, y_train) #training model
```

```python
c=0
for i in range(y_test.shape[0]):
    p=model.predict(x_test[i])
    if p==y_test[i]:
        c=c+1;
accuracy=(c*100)/(y_test.shape[0]) #testing model and calculating accuracy
```

```python
print("Accuracy is ", accuracy)
```
